```python
#Creating variable to store the page numbers to target for each specification
p=[0]*36
pp=[]
for x in range(56):
    pp.append(p)
pcount=pd.DataFrame(pp)
pcount
```

```python
# Storing the actual page numbers
# takes -1 if no registerd starup so that we can skip reading that page later
for i in range (0,56):
    for j in range (0,36):
        if scount.iloc[i][j] == 0:
            pcount.iloc[i][j]=-1
        elif scount.iloc[i][j] < 10:
            pcount.iloc[i][j]=0
        elif scount.iloc[i][j]<19:
            pcount.iloc[i][j]=1
        elif scount.iloc[i][j]<28:
            pcount.iloc[i][j]=2
        elif scount.iloc[i][j]<37:
            pcount.iloc[i][j]=3
        elif scount.iloc[i][j]<46:
            pcount.iloc[i][j]=4
        elif scount.iloc[i][j]<55:
            pcount.iloc[i][j]=5
        elif scount.iloc[i][j]<64:
            pcount.iloc[i][j]=6
        elif scount.iloc[i][j]<73:
            pcount.iloc[i][j]=7
        elif scount.iloc[i][j]<82:
            pcount.iloc[i][j]=8
        elif scount.iloc[i][j]<91:
            pcount.iloc[i][j]=9
        elif scount.iloc[i][j]<100:
            pcount.iloc[i][j]=10
        elif scount.iloc[i][j]<109:
            pcount.iloc[i][j]=11
        else:
            z=scount.iloc[i][j]
            zz=int(z/9)-2
            pcount.iloc[i][j]=zz

pcount
```

```python
#Codes to get startup information and store it
# Here, for 'i' it must go on till len(industries)
#'j' it must go on till len(states)
# Because of time out error we have extarcted the data in different batches


data_store=[]
for i in range (0,1):
    for j in range (0,1):
        if pcount.iloc[i][j]==-1:
            continue
        else :
            url=url1+industries[i]+url2+states[j]+url3+str(pcount[i][j])
            options = webdriver.ChromeOptions()
            options.set_capability("loggingPrefs", {'performance': 'ALL'})
            options.add_experimental_option("excludeSwitches", ['enable-automation'])
            service = ChromeService(executable_path=r"C:\Users\HP\Downloads\chromedriver.exe")
            driver = webdriver.Chrome(service=service, options=options)
            driver.get(url)
            time.sleep(40)
            soup = BeautifulSoup(driver.page_source,'lxml')
            data=pd.DataFrame(columns=['Name','Stage','Location','Sector'])
            startups=soup.find_all('div',class_="col-md-4 col-sm-6 col-space20")
            for startup in startups:
                Sup_name_loc=startup.find('div',class_="events-details")
                Sup_name=Sup_name_loc.h3.text
                Sup_stage_loc=startup.find('span',class_="highlighted-text")
                Sup_stage=Sup_stage_loc.text
                Sup_location_loc=startup.find('li',class_="location")
                Sup_location=Sup_location_loc.span.find_next_siblings("span")
                Sup_sector_loc=startup.find('span',class_="dept")
                Sup_sector=Sup_sector_loc.text
                Sup_sector=Sup_sector.lstrip()
                Sup_sector=Sup_sector.rstrip()
                Sup_link_loc=startup.find('a',class_="img-wrap")
                x=str(Sup_link_loc)
                xs=x.find('href=')
                xe=x.find('target')
                data=data.append({'Name':Sup_name, 'Stage':Sup_stage, 'Location':Sup_location,
                      'Sector':Sup_sector,'Link':x[xs+6:xe-2]}, ignore_index = True)
            data_store.append(data)
            print(data.iloc[0][2],"-",data.iloc[0][3],"-",len(data))
            #print(startups.find_all("span",class_="total"))
            driver.close()
```

```python
#Combining all the startups details into a single frame
for i in range(0,len(data_store)-1):
    data_store[0]=data_store[0].append(data_store[i+1],ignore_index=True)
```
